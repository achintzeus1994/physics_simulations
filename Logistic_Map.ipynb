```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## I want to plot logistic map
r = 3.55
x = np.linspace(0,1,100)
y = r*x*(1-x)
def logistic(xin):
    yin=r*xin*(1-xin)
    return yin
```

```python
niter =250
x0 = np.random.rand()
px = np.zeros(niter+1)
py = np.zeros(niter+1)
px[0]=x0;
for i in range(0,niter-1,2):
    px[i+1]=px[i]
    py[i+1] = logistic(px[i])
    px[i+2]=py[i+1]
    py[i+2]=py[i+1]
```

```python
plt.plot(x,y,label='$y=rx(1-x)$')
plt.plot(x,x,label='$y=x$')
plt.plot(px,py,c='b',alpha=0.7)
plt.xlim(0,1);plt.ylim(0,1)
plt.xlabel('x '); plt.ylabel('$r\,x\,(1-x)$');plt.grid(True)
plt.title('$r={:},x0={:.1}$'.format(r,xiter[0]))
plt.legend()
plt.show()
```
